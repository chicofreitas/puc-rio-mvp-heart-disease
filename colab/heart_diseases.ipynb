## 1 Definição do Problema

O dataset utilizado será o *Heart Disease Dataset* fornecido pelo *UCI Machine Learning Repository*. Seu objetivo é classificar os pacientes em grupos onde **não possuem** doenças cardíacas, ou **possuem alguma** das quatro doenças cardíacas registradas com base em dados médicos disponíveis nos datasets. Este dataset é formado por um conjunto de quatro bases de dados de diferentes localidades: *Cleveland, Hungary, Switzerland, e o VA Long Beach*.

O dataset apresenta em diversos atributos relacionados a dados médicos e uma variável de classe (*num*) não binária representada pelos valores 0 (ausêcia de doença cardíaca) e 1 a 4 para indicar a presença de alguma dentre quatro doenças cardíacas.

## Este banco de dados contém 76 atributos, mas todos os experimentos publicados referem-se ao uso de um subconjunto de 14 deles. 

Em particular, o banco de dados de Cleveland é o único que foi usado por pesquisadores de ML até hoje. Os experimentos com o banco de dados Cleveland se concentraram em simplesmente tentar distinguir a presença (valores 1,2,3,4) da ausência (valor 0).

### Sobre este Sprint

A pretensão deste projeto é unir os quatro datasets, ao invés de utilizar apenas um, para realizar as devidas análises e pré-processamentos para o treinamento futuro de modelos de *ML*.

Consulte https://archive-beta.ics.uci.edu/dataset/45/heart+disease para mais informações sobre os datasets e o dowlnoad de seus arquivos.

# 2 Carga dos Dados

Os quatro datasets foram transformados para o formato CSV e armazenados no repositório https://github.com/chicofreitas/puc-rio-mvp-heart-disease, dentro do diretório */resources/datasets/heart-disease/csv*.

Algumas linhas do dataset de **Cleveland** não possuem a mesma quantidade de atributos dos demais registros, o que dificulta a leitura do arquivo *CSV* utilizando o Pandas. Para manter uma integridade do conjunto de dados de Cleveland, apenas 282 linhas tiveram que ser lidas (note o parâmetro **nrows** na leitura do dataset de Cleveland), tendo em vista que apenas as útimas linhas apresentaram este erro. 

Em seguida criamos uma lista com os dataframes, e utilizamos a função *concat()* do Pandas para concatenar os quatro datasets. Desta forma, temos um dataset com mais dados para análise e futuros treinamentos de modelos de *ML*.

```python
cleveland_dataset = 'https://github.com/chicofreitas/puc-rio-mvp-heart-disease/raw/main/resources/datasets/heart-disease/csv/Cleveland/cleveland.csv'
hungarian_dataset = 'https://github.com/chicofreitas/puc-rio-mvp-heart-disease/raw/main/resources/datasets/heart-disease/csv/Hungarian/hungarian.csv'
long_beach_dataset = 'https://github.com/chicofreitas/puc-rio-mvp-heart-disease/raw/main/resources/datasets/heart-disease/csv/Long-Beach/long-beach-va.csv'
switzerland_dataset = 'https://github.com/chicofreitas/puc-rio-mvp-heart-disease/raw/main/resources/datasets/heart-disease/csv/Switzerland/switzerland.csv'

hd_c = pd.read_csv(cleveland_dataset, names=labels_attributes, on_bad_lines='skip', nrows=282)
hd_h = pd.read_csv(hungarian_dataset, names=labels_attributes, on_bad_lines='skip')
hd_l = pd.read_csv(long_beach_dataset, names=labels_attributes, on_bad_lines='skip')
hd_s = pd.read_csv(switzerland_dataset, names=labels_attributes, on_bad_lines='skip')

data_frames = [hd_c, hd_h, hd_l, hd_s]

df = pd.concat(data_frames)
```
